```python
from datetime import datetime
from datetime import date

# Splitting the dataframe into 2: date column labels and obj labels
Confirmed1 = jh_Confirmed.loc[:, jh_Confirmed.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]
#print (Confirmed1)
Confirmed2 = jh_Confirmed.loc[:, ~jh_Confirmed.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]

Recovered1 = jh_Recovered.loc[:, jh_Recovered.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]
Recovered2 = jh_Recovered.loc[:, ~jh_Recovered.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]


Death1 = jh_Death.loc[:, jh_Death.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]
Death2 = jh_Death.loc[:, ~jh_Death.columns.isin (['Province/State', 'Country/Region','Lat' ,'Long'])]


# Convert the date column labels to true datetime 
Confirmed2.columns = pd.to_datetime(Confirmed2.columns, format='%m/%d/%y').date
#print(Confirmed2)
Recovered2.columns = pd.to_datetime(Recovered2.columns, format='%m/%d/%y').date
Death2.columns = pd.to_datetime(Death2.columns, format='%m/%d/%y').date


#recombine the splitted dataframes
display("Recombined Dataframes")
jh_Confirmed = Confirmed1.join(Confirmed2)
jh_Recovered = Recovered1.join(Recovered2)
jh_Death = Death1.join(Death2)

#print(jh_Confirmed)
jh_Confirmed.info() 
print("\n")
jh_Recovered.info()
print("\n")
jh_Death.info()


#datetime.strptime("df2.columns.values", '%m/%d/%y').strftime("%M/%d/%Y")
#df2.astype({'data': 'datetime64[ns]'})
#data = df2.columns.values.astype(datetime) 
```

Determine the effect of the Pandemic on the economy. To do this, you will compare the Real GDP value Pre-COVID-19 with Real GDP in 2020 (COVID-19 Period, especially Q2 2020)

Steps
* From the Real GDP Data, generate a `barplot` using the GDP values for each year & quarters. For example: On x-axis you will have year 2017 and the bars will be values of each quarters(Q1-Q4). You expected to have subplots of each quarters on one graph.

Hint: Use [Pandas.melt](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to create your plot DataFrame

```python
RealGDP = pd.read_csv("RealGDP.csv")
data1 = pd.melt(RealGDP, id_vars =['Year'], value_vars =['Q1'])
data2 = pd.melt(RealGDP, id_vars =['Year'], value_vars =['Q2'])
data3 = pd.melt(RealGDP, id_vars =['Year'], value_vars =['Q3'])
data4 = pd.melt(RealGDP, id_vars =['Year'], value_vars =['Q4'])

#Using subplot without sharing y axis
fig = plt.figure(figsize=(8, 8))

fig.suptitle('A plot of Overall CCVI Index and Confirmed Cases')
ax1 = fig.add_subplot(411)
plt.bar(data1['Year'], data1["value"], align='center', color='b')

ax2 = fig.add_subplot(412)
plt.bar(data2['Year'], data2['value'], align='center', color='g')
point = data2['value'][data2['Year'] == 2020 ]  #for the line on Q2, 2020
plt.axhline(y =point.iloc[0] , color ="red", linestyle ="--")

ax3 = fig.add_subplot(413)
plt.bar(data3['Year'], data3["value"], align='center', color='k')

ax4 = fig.add_subplot(414)
plt.bar(data4['Year'], data4['value'], align='center', color='r')

# Hide x labels and tick labels for all but bottom plot.
for ax in [ax1, ax2, ax3]:
    ax.label_outer()

ax1.title.set_text('Q1')
ax2.title.set_text('Q2')
ax3.title.set_text('Q3')
ax4.title.set_text('Q4')
plt.show()

```

```python
# plot grouped bar chart
RealGDP.plot(x='Year', kind='bar', stacked=False, title='Comparative quarterly plot of GDP')
point = data2["value"][data2['Year'] == 2020 ]  #for the line on Q2, 2020
plt.axhline(y =point.iloc[0] , color ="red", linestyle ="--", label = "Trace line")
plt.legend(loc='lower left')

```

- The quarterly barplots show that there is a decrease in the Real GDP for Q2 2020 as compared with Q2 of previous years as a result of the emergence of Covid 19
- Likewise for Q3 2020, there is a decrease when compared with Q3 2019
